````python
class Encryption:
    def command_prompt(self):
        self.key = 0
        Encryption.set_key(self)
        while True: ## infinite loop
            cmd = input('Enter your command here: > ')
            if cmd == 'show message':
                Encryption.show_message()
                continue
            elif cmd == 'show encrypted':
                Encryption.show_encrypted()
                continue
            elif cmd == 'show key':
                Encryption.show_key(self)
                continue
            elif cmd == 'set key':
                Encryption.set_key(self)
                continue
            elif cmd == 'show map':
                Encryption.show_map(self)
                continue
            elif cmd == 'encrypt':
                Encryption.encrypt(self)
                continue
            elif cmd == 'decrypt':
                Encryption.decrypt(self)
                continue
            elif cmd == 'bye':
                print('Bye! Have a nice day.')
                break
            else:
                print('Error. Invalid command.')
                continue
    
    def show_message():
        try:
            f = open('message.txt','r')
            for i in f:
                print(i)
            f.close()
        except:
            print('Error. \'message.txt\' is not available.')
    
    def show_encrypted():
        try:
            f = open('encrypted.txt','r')
            for i in f:
                print(i)
            f.close()
        except:
            print('Error. \'encrypted.txt\' is not available.')
    
    def show_key(self):
        print('The encryption key is: ' + str(self.key)) 
        
    def set_key(self):
        inputkey = input('Type the encryption key: ')
        try:
            self.key = int(inputkey)
            if (self.key > 999) or (self.key < 100):
                print('Error: it should be a number between 100 and 999.')
                Encryption.set_key(self)
            print('Encryption key successfully set as ' + str(self.key))
            self.key = self.key
        except:
            print('Please type in an integer.')
            Encryption.set_key(self)
            
    def show_map(self):
        if self.key == 123:
            import string 
            alphabet = string.ascii_lowercase[:]
            l1 = list(alphabet)
            l2 = ['e','g','l','j','h','o','t','u','p','v','f','s','x','a','w','q','k','r','m','z','d','y','n','c','i','b']
            self.encryption = {}
            for key in l1:
                for value in l2:
                    self.encryption[key] = value
                    l2.remove(value)
                    break
            print("The encryption map is:")
            for key, value in self.encryption.items():
                print( key + ' => ' + value, end = '\t')
            self.decryption = {}
            l1 = list(alphabet)
            l2 = ['e','g','l','j','h','o','t','u','p','v','f','s','x','a','w','q','k','r','m','z','d','y','n','c','i','b']
            for key in l2:
                for value in l1:
                    self.decryption[key] = value
                    l1.remove(value)
                    break
            print("\nAnd The decryption map is:")
            for key, value in self.decryption.items():
                print( key + ' => ' + value, end = '\t')
            print('\n')
        else:
            print('Wrong key, try set up a new key')
            
    def encrypt(self):
        try:
            my_file = open('encrypted.txt','w')
            file = open('message.txt','r')
            text = ''
            for i in file:
                 text = text + i
            file.close()
            list1 = list(text)
            new_l = []
            for i in list1:
                if i in self.encryption:
                    new_l.append(self.encryption[i])
                else:
                    new_l.append(i)  
            my_file.write(''.join(new_l))
            my_file.close()
            print('Successful encryption.') 
        except:
            print('Error. The encryption failed.')
            
    def decrypt(self):
        try:
            file = open('encrypted.txt','r')
            text = ''
            for i in file:
                text = text + i
            file.close()
            list2 = list(text)
            new_l = []
            for i in list2:
                if i in self.decryption:
                    new_l.append(self.decryption[i])
                else:
                    new_l.append(i)
            print('Successfully decrypted.\nYour message is:')
            print('\n' + ''.join(new_l))
        except:
            print('Error. The decryption failed.')
````
